```python
from Methods.patchExtractionFunctions import divergence_date
from Methods.patchExtractionFunctions import pr_patches

# Sample diverged variants
mainline = 'apache/kafka'  # ml
fork = 'linkedin/kafka'  # fv

ct = 0  # ct = token counter
fork_date, divergence_date, least_date, ahead_by, behind_by, ct = divergence_date(mainline, fork, ct)
print(fork_date, divergence_date, least_date, ahead_by, behind_by, ct)


# The diverge_date and least_date can be changed to narrow the time interval for the merged pull requests
# divergence_date = '2021-12-06T16:47:56Z'
# least_date = '2022-02-24T08:03:42Z'

pr_patch_ml_str = ''
pr_patch_fv_str = ''
pr_title_ml_str = ''
pr_title_fv_str = ''

pr_patch_ml, pr_title_ml, pr_all_merged_ml, ct = pr_patches(mainline, divergence_date, least_date, ct)
pr_patch_fv, pr_title_fv, pr_all_merged_fv, ct = pr_patches(fork, divergence_date, least_date, ct)

# at least one of the mainline or fork should have a pr with patch
if len(pr_patch_ml) > 0 or len(pr_patch_fv) > 0:
    if len(pr_patch_ml) > 1:
        pr_patch_ml_str = '/'.join(map(str, pr_patch_ml))
        pr_title_ml_str = '=/='.join(map(str, pr_title_ml))
    if len(pr_patch_ml) == 1:
        pr_patch_ml_str = pr_patch_ml[0]
        pr_title_ml_str = pr_title_ml[0]

    if len(pr_patch_fv) > 1:
        pr_patch_fv_str = '/'.join(map(str, pr_patch_fv))
        pr_title_fv_str = '=/='.join(map(str, pr_title_fv))
    if len(pr_patch_fv) == 1:
        pr_patch_fv_str = pr_patch_fv[0]
        pr_title_fv_str = pr_title_fv[0]

print('mainline', mainline)
print('fork', fork)
print('divergence_date', divergence_date)
print('least_date', least_date)
print('fork unique commits', ahead_by)
print('mainline unique commits', behind_by)

print('\nmainline patch details')
print('=======================')
print('number of merged pr = ', len(pr_all_merged_ml))
print('number of patches =', len(pr_patch_ml))
print('patches::', pr_patch_ml_str)
print('patches titles::', pr_title_ml_str)

print('\nfork patch details')
print('=======================')
print('number of merged pr =', len(pr_all_merged_fv))
print('number of patches =', len(pr_patch_fv))
print('patches ::', pr_patch_fv_str)
print('patches titles::', pr_title_fv_str)
```
